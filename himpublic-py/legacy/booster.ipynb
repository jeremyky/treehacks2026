```python
booster.enable_arm_usage()
```

```python
from booster_python_client.lib import get_joint_name_by_index
import booster_python_client.joints as joint_lists
from booster_python_client.helpers import tuple_to_joint_dict


## Record robot position
filename = "q_values.txt"
with open(filename, "w") as f:
    f.write(", ".join([get_joint_name_by_index(j) for j in joint_lists.LEFT_RIGHT_ARM_TORSO_JOINTS_INDICES]) + "\n")

while True:
    input("Press Enter to record again...")
    q_str = booster.read_latest_low_state(joint_lists.LEFT_RIGHT_ARM_TORSO_JOINTS_INDICES)
    with open(filename, "a") as f:
        f.write(q_str + ",\n")
```

```python
## PASTE JOINT VALUES HERE
PUNCH = [
 (-0.6117, -1.4550, 0.7559, -1.7103, -0.3317, 1.2314, 0.5152, 2.0235),
(-0.8948, -1.4565, 0.7559, -1.7088, -0.5934, 1.2314, 0.5190, 1.9774),
(-1.0985, -1.4527, 0.7776, -1.6356, -0.8177, 1.2791, 0.5152, 1.9282),
(-1.3457, -1.4157, 0.7197, -1.6341, -1.1557, 1.2810, 0.5789, 1.8282),
]

for i in range(len(PUNCH)):
    PUNCH[i] = tuple_to_joint_dict(
        joint_lists.LEFT_RIGHT_ARM_TORSO_JOINTS_INDICES, PUNCH[i]
    )
```

```python
## TEST THE ACTION
booster.send_command(PUNCH, speed="slow", time_gap_s=0.5)
```
